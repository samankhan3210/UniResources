```python
month_names = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep","Oct", "Nov", "Dec"]
months = range(12)
conversion = [0.05, 0.08, 0.18, 0.28, 0.4, 0.66, 0.74, 0.78, 0.8, 0.81, 0.85, 0.85]
plt.xlabel("Months")
plt.ylabel("Conversion")
plt.plot(months, conversion)
ax=plt.subplot()
ax.set_xticks(months)
ax.set_xticklabels(month_names)
ax.set_yticks([0.1, 0.25, 0.5, 0.75])
ax.set_yticklabels(['10%', '25%', '50%', '75%'])
plt.show()
```

```python
import numpy as np
drinks = ["cappuccino", "latte", "chai", "americano", "mocha", "espresso"]
sales1 = [91, 76, 56, 66, 52, 27]
sales2 = [65, 82, 36, 68, 38, 40]
x = np.arange(len(drinks))
width = 0.35
fig, ax = plt.subplots()
rects1 = ax.bar(x - width / 2, sales1, width, label='Branch1')
rects2 = ax.bar(x + width / 2, sales2, width, label='Branch2')
ax.set_ylabel('Sales')
ax.set_title('Sales by Branches and Drinks')
ax.set_xticks(x)
ax.set_xticklabels(drinks)
ax.legend()
plt.show()
```

```python
drinks = ["cappuccino", "latte", "chai", "americano", "mocha", "espresso"]
sales1 = [91, 76, 56, 66, 52, 27]
sales2 = [65, 82, 36, 68, 38, 40]
plt.bar(range(len(drinks)), sales1)
plt.bar(range(len(drinks)), sales2, bottom=sales1)
plt.legend(['Location1', 'Location2'], loc=1)
plt.show()
```
