```python
# Question number 2 a
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import  train_test_split,KFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

df = pd.read_csv('glass.csv')
print(df.head(4))
arr =  df.values
Y =  arr[:,9]
df.drop(['Type'], axis=1, inplace=True)
scaler = StandardScaler()
data_scaled = scaler.fit_transform(df)
df2 = pd.DataFrame(data_scaled)

print(df2.describe)
arr =  df2.values
X =  arr[:,0:8]
acc = []
kn = []
for k in range(1,6):
    knn = KNeighborsClassifier(metric='minkowski', n_neighbors=k)
    x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.3)
    knn.fit(x_train, y_train)
    pred = knn.predict(x_test)
    acc.append(accuracy_score(y_test,pred))
    kn.append(k)
    
for i in range(len(acc)):
    for j in range(len(acc)):
        if acc[i] > acc[j]:
            temp = acc[i]
            acc[i] = acc[j]
            acc[j] = temp
            temp2 = kn[i]
            kn[i] = kn[j]
            kn[j] = temp2

print("Highest Accuracy = ", acc[0], "for k = ", kn[0])
```

```python
# Question number 3 a
import numpy as np
import pandas as pd

from sklearn.metrics import mean_squared_error
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import KFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.metrics import SCORERS

df = pd.read_excel('energy_efficiency.xlsx')
df.drop(df.columns[[10,11]], axis=1, inplace=True)
print(df.head(4))
scaler = StandardScaler()
data_scaled = scaler.fit_transform(df)
print(pd.DataFrame(data_scaled).describe)
df2 = pd.DataFrame(data_scaled, columns=['X1', 'X2', 'X3', 'X4', 'X5','X6','X7','X8','Y1', 'Y2'])
print(df2.head(4))
df2.drop(['X2','X5','X7','X8'], axis=1, inplace=True)
print(df2.head(4))
predictors = ['X1', 'X3', 'X4', 'X6']
x = df2[predictors]
y1 = df2['Y1']
y2 = df2['Y2']
print(x)
print(y2)
print(y1)
print("\n\nFor Y1\n\n")
kf = KFold(n_splits=5, shuffle=True)
lm = LinearRegression()
scores = cross_val_score(lm, x, y1, scoring='neg-mean-squared-error', cv=kf, n_jobs=-1)
print("Accuracy = ", np.sqrt((scores)))
#print("Root mean square error =", np.sqrt(mean_squared_error(scores)))

print("\n\nFor Y2\n\n")
kf = KFold(n_splits=5, shuffle=True)
lm = LinearRegression()
scores = cross_val_score(lm, x, y2, scoring='accuracy', cv=kf, n_jobs=-1)
#print("Root mean square error =", np.sqrt(mean_squared_error(scores)))
print("Accuracy = ", np.mean(scores))
```
