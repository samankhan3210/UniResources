```python
import matplotlib.pyplot as plt
import numpy as np

heights = np.random.normal(160, 5, 250)
plt.hist(heights)
plt.title("Heights of 250 people")
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

age = [10, 20, 30, 40, 50]
weight = [35, 45, 55, 65, 75]
plt.scatter(age,weight)
plt.title("Disadvantages of Smoking with Increasing Age")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.show()
```
