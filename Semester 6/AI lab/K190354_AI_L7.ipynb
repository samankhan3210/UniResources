```python
# Question number 01

player_x = 'x'
player_o = 'o'

def evaluate(b):
    for row in range(3):
        if (b[row][0] == b[row][1] and b[row][1] == b[row][2]):
            if (b[row][0] == player_x):
                return 10
            elif (b[row][0] == player_o):
                return -10

    for col in range(3):
        if (b[0][col] == b[1][col] and b[1][col] == b[2][col]):
            if (b[0][col] == player_x):
                return 10
            elif (b[0][col] == player_o):
                return -10

    if (b[0][0] == b[1][1] and b[1][1] == b[2][2]):
        if (b[0][0] == player_x):
            return 10
        elif (b[0][0] == player_o):
            return -10

    if (b[0][2] == b[1][1] and b[1][1] == b[2][0]):
        if (b[0][2] == player_x):
            return 10
        elif (b[0][2] == player_o):
            return -10

    return 0


def minimax(board, depth, isMax):
    score = evaluate(board)
    if (score == 10):
        return score

    if (score == -10):
        return score

    for i in range(3):
        for j in range(3):
            if (board[i][j] == '_'):
                return 0

    if (isMax):
        best = -1000
        for i in range(3):
            for j in range(3):
                if (board[i][j] == '_'):
                    board[i][j] = player_x
                    best = max(best, minimax(board,depth + 1,not isMax))
                    board[i][j] = '_'
#         print(board)
        return best

  
    else:
        best = 1000
#         print(board)
        for i in range(3):
            for j in range(3):
                if (board[i][j] == '_'):
                    board[i][j] = player_o
                    best = min(best, minimax(board, depth + 1, not isMax))
                    board[i][j] = '_'
#         print(board)
        return best


def findoptMov(board):
    bestVal = -1000
    optMov = (-1, -1)
    for i in range(3):
        for j in range(3):
            if (board[i][j] == '_'):
                board[i][j] = player_x
                moveVal = minimax(board, 0, False)
                board[i][j] = '_'
                if (moveVal > bestVal):
                    optMov = (i, j)
                    bestVal = moveVal
    print("The value of the best Move is :", bestVal)
    print()
    return optMov


board = [['o', '_', 'x'],
         ['x', '_', '_'],
         ['x', 'o', 'o']]

optMov = findoptMov(board)
print("The Optimal Move is : (" , optMov[0], ",", optMov[1], ")")
```

```python
# Question number 02

MAX = 1000
MIN = -1000

def minimax(depth, nodeIndex, maximizingPlayer,values, alpha, beta):
    if depth == 3:
        return values[nodeIndex]
 
    if maximizingPlayer:
        best = MIN
        for i in range(0, 2):
            val = minimax(depth + 1, nodeIndex * 2 + i,False, values, alpha, beta)
            best = max(best, val)
            alpha = max(alpha, best)
            if beta <= alpha:
                break
        return best
      
    else:
        best = MAX
        for i in range(0, 2):
            val = minimax(depth + 1, nodeIndex * 2 + i,True, values, alpha, beta)
            best = min(best, val)
            beta = min(beta, best)
            if beta <= alpha:
                break
          
        return best
      
        
values = [2,4,6,8,1,2,10,12] 
print("The optimal value is :", minimax(0, 0, True, values, MIN, MAX))
```

```python
# Question number 03

global N
N = 4
        
def isSafe(board, row, col):
    for i in range(col):
        if board[row][i] == 'Q':
            return False
 
    for i, j in zip(range(row, -1, -1), range(col, -1, -1)):
        if board[i][j] == 'Q':
            return False
 
    for i, j in zip(range(row, N, 1), range(col, -1, -1)):
        if board[i][j] == 'Q':
            return False
    return True
 
def NQ(board, col):
    if col >= N:
        return True

    for i in range(N):
        if isSafe(board, i, col):
            board[i][col] = 'Q'
            if NQ(board, col + 1) == True:
                return True
            board[i][col] = "_"
    return False


board = [["_", "_", "_", "_"],
         ["_", "_", "_", "_"],
         ["_", "_", "_", "_"],
         ["_", "_", "_", "_"]]

if NQ(board, 0) == False:
    print ("Solution does not exist")

for i in range(N):
    for j in range(N):
        print (board[i][j],end=' ')
    print()
```

```python
# Question number 04

import itertools
def get_value(word, substitution):
    s = 0
    factor = 1
    for letter in reversed(word):
        s += factor * substitution[letter]
        factor *= 10
    return s


def solve2(equation):
    left, right = equation.lower().replace(' ', '').split('=')
    left = left.split('+')
    letters = set(right)
    for word in left:
        for letter in word:
            letters.add(letter)
    letters = list(letters)

    digits = range(10)
    for perm in itertools.permutations(digits, len(letters)):
        sol = dict(zip(letters, perm))

        if sum(get_value(word, sol) for word in left) == get_value(right, sol):
            print( "{}".format(sol))
            

solve2("BASE + BALL  = GAMES")
```
