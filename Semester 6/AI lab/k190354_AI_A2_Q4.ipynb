```python
# Breadth First Search
def BFS(src, dst):
    queue = []
    visited = []
    bfs_cost = {source : 0}
    parent = {source : None}
    queue.append(source)
    visited.append(source)
    print("Path : ", end = " ")
    while queue:
        current = queue.pop(0)
        print(current, end = "  ") 
        for neighbour in lst[current]:
            if neighbour not in visited:
                visited.append(neighbour)
                queue.append(neighbour)
                bfs_cost[neighbour] = bfs_cost[current] + lst[current][neighbour]['weight']
                parent[neighbour] = current

        if current == destination:
            print("\nCost = ",bfs_cost[destination])
            break
```

```python
# Uniform Cost Search
def UCS(src, dst):
    Total_Cost = 0
    visited = []
    Queue = PriorityQueue()
    Queue.put((0,src))
    parent = {src : None}
    while Queue:
        cost, current = Queue.get()
        if current not in visited:
            visited.append(current)
            if current == dst:
                Total_Cost = cost
                break
            for neighbour in lst[current]:
                if neighbour not in visited:
                    Total_Cost = cost + lst[current][neighbour]['weight']
                    Queue.put((Total_Cost,neighbour))
                    parent[neighbour] = current

    path = [dst]
    temp = dst
    while parent[temp] != src:
        path.append(parent[temp])
        temp = parent[temp]
    path.append(src)
    path.reverse()

    print("Path : ",path)
    print("Cost = ", Total_Cost)  
```

```python
# Greedy Best First Search
def GBFS(src, dst):
    Queue = PriorityQueue()
    Queue.put((0,src))
    Total_cost = 0
    Visited = []
    print("Path : ", end = " ")
    while Queue:
        cost, current = Queue.get()
        print(current, end="  ")
        Total_cost += cost
        if current == dst:
            break
        else:
            for neighbour in lst[current]:
                if neighbour not in Visited:
                    Visited.append(neighbour)
                    Queue.put((h_c[neighbour],neighbour))

    print("\nCost = ", Total_cost)
```

```python
#Iterative Deepening Depth First Search
Total_cost = 0
Total_path = []

def DepthLimited(src,dst,limit):
    if src == dst :
        Total_path.append(src)
        return True 
    
    if limit < 0 :
        return False
    
    for neighbour in lst[src]:
        if DepthLimited(neighbour,dst,limit-1):
            Total_path.append(src)
            
            return True 
        
    return False


def IDDFS(src,dst,limit):
    print("Path : " , end=" ") 
    for x in range(limit):
        if DepthLimited(src,dst,x) == True :
            Total_path.reverse()
            for i in Total_path:
                print(i,end="  ")
            return
    return("Path not found") 
```

```python
print("Path way and cost for BFS : ")
BFS(source, destination)
print("Path way and cost for UCS : ")
UCS(source, destination)
print("Path way and cost for GBFS : ")
GBFS(source, destination)
print("Path way for IDDFS : ")
IDDFS(source,destination,5)
```
